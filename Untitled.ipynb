```python
 import ast

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L   
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python

def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter <3:
            L.append(i['name'])
            counter+=1 
        else:
            break
    return L   
```

```python
movie['cast'].apply(convert3)
```

```python
movie['cast']=movie['cast'].apply(convert3)
```

```python
def Director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            L.append(i['name'])
            break
    return L   
```

```python
movie['crew'].apply(Director)
```

```python
movie['crew']=movie['crew'].apply(Director)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps =PorterStemmer()
```

```python
def stem_fun(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)   
```

```python
movie['tag'].apply(stem_fun)
```

```python
movie['tag'] =movie['tag'].apply(stem_fun)
```

```python
distance = similarity[3]
movies_list = sorted(list(enumerate(distance)),reverse=True,key = lambda x:x[1])[1:11]
for i in movies_list:
    print(movie['title'][i[0]])
```

```python
def recommend(film):
    mov_index = movie[movie['title']==film].index[0]
    
    distance = similarity[mov_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key = lambda x:x[1])[1:11]
    for i in movies_list:
        print(movie['title'][i[0]])
```

```python
recommend("Batman")
```
